```python
import pandas as pd
import math


def read_data():
    factorize_dict = list()
    data = pd.read_csv('data_set.csv',
                       encoding='UTF-8',
                       sep=';')
    del data['Имя']
    to_enumerate = ['Пол', 'Высшая школа', 'Округ', 'Цвет глаз', 'К/ч']
    for column in data.columns:
        factorize_dict.append(list(pd.factorize(data[column])[1]))
    data[to_enumerate] = data[to_enumerate].apply(lambda x: pd.factorize(x)[0])
    data = data.apply(lambda x: [elem / max(x) for elem in x])
    return data, factorize_dict

data, factorize_dict = read_data()

def sort_neighbours_by_distance(entry):
    neighbours_distance = []
    for index, value in data.iterrows():
        if index != entry:
            coordinates_difference = [x - y for x in data.iloc[entry][:-1] for y in value[:-1]]
            distance = math.sqrt(sum(x ** 2 for x in coordinates_difference))
            neighbours_distance.append([distance, value[-1]])
    return sorted(neighbours_distance)


def predict_by_k_neighbours(k, entry):
    closest_neighbours = sort_neighbours_by_distance(entry)[:k+1]
    prediction = round(sum(x[1] for x in closest_neighbours) / k)
    return prediction == data.iloc[entry][-1]


def find_optimal_k():
    correct_guesses = []
    for k in range(1, len(data)-1):
        guesses = []
        for entry in range(1, len(data)):
            guesses.append(predict_by_k_neighbours(k, entry))
        correct_guesses.append(sum(guesses))
    return correct_guesses.index(max(correct_guesses)), correct_guesses


print(find_optimal_k())
```
